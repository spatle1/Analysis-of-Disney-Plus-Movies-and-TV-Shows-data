```python
Make a copy of orginal data and remove unwanted column for task 2(Realation between type and rating)
```

```python
disney_plus_df_copy = df.copy()
disney_plus_df = df.copy()

disney_plus_df_copy.drop(['director','cast','date_added','marvel_t'],axis=1,inplace=True)
disney_plus_df_copy
```

```python
Task 2: Releation between Type(movies or TV shows) and rating
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x='rating',hue='type',data=disney_plus_df_copy)
plt.title('Relation between Type and Rating of Movies and TV Shows')
plt.show();
```

```python
TASK 3. data segregation for different country as which country produce most number of shows and movies
```

```python
Countries_data = disney_plus_df_copy.groupby('country').count().sort_values('title',ascending=False).head(10)

Countries_data.reset_index(inplace=True)
Countries_data
plt.bar(Countries_data.country,Countries_data.title);
```

```python
TASK 4. Top 10 director based on most title for movies and tv shows
```

```python
director_with_most_moviesshows_title = disney_plus_df.groupby('director').count().sort_values('rating',ascending=False).head(10)
director_with_most_moviesshows_title.reset_index(inplace=True)
plt.bar(director_with_most_moviesshows_title.director,director_with_most_moviesshows_title.title);
```
